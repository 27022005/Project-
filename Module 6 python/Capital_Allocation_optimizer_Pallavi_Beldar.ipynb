```python
df =data.dropna()
df
```

```python
print(df.isnull().sum())
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=True)
```

```python
print(df['Date'])
```

```python
selected = ['NYA', 'IXIC', 'HSI', 'N225']

df = df[df['Index'].isin(selected)]
df
```

```python
df['Daily Return'] = df.groupby('Index')['Close'].pct_change()
```

```python
volatility = df.groupby('Index')['Daily Return'].std()
volatility
```

```python
avg_return = df.groupby('Index')['Daily Return'].mean()
```

```python
summary = pd.DataFrame({
    'Average Return': avg_return,
    'Volatility': volatility
})
```

```python
summary['Risk Score'] = summary['Average Return'] / summary['Volatility']
```

```python
total_capital = 20000000
```

```python
summary['Weight'] = summary['Risk Score'] / summary['Risk Score'].sum()
```

```python
summary['Capital Allocation'] = summary['Weight'] * total_capital
```

```python
summary = summary.sort_values(by='Risk Score', ascending=False)
```

```python
print(summary)
```

```python
import matplotlib.pyplot as plt

plt.bar(summary.index, summary['Capital Allocation'])
plt.title('Capital Allocation')
plt.xlabel('Strategy (Index)')
plt.ylabel('Capital')
plt.show()
```

```python
summary.to_csv('Capital Allocation Optimizer.csv')
```
