```python
plt.figure(figsize=(8,5))
P = sns.histplot(df["Temperature"], kde=True,color="#56095D")
P.set_facecolor("#E9A6DE")   
plt.title(f"Distribution of Temperature")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Season', y='Temperature', data=df)
plt.title("Temperature by Season")
plt.show()
```

```python
agg = df.groupby('Cloud Cover')['Visibility (km)'].sum()

percent = (agg / agg.sum()) * 100

plt.figure(figsize=(7,7))
plt.pie(percent, labels=percent.index, autopct='%1.1f%%')
plt.title('Cloud Cover Wise Total Visibility Percentage')
plt.show()
```

```python

df = df[(df["Temperature"] > -10) & (df["Temperature"] < 50)]

plt.figure()
plt.scatter(df["Temperature"],df["Humidity"],alpha=0.4,s=20)
plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity (Clear View)")
plt.show()
```

```python
plt.figure(figsize=(50,50))
sns.catplot(data=df,x="Location",y="UV Index",hue="Weather Type",
            col="Season",kind="bar",height=5,aspect=1)
plt.subplots_adjust(top=0.85)
plt.suptitle("UV Index by Location + Season + Weather Type")
plt.show()
```
